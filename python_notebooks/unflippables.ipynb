# Unfound states
Some states are not found by the repeated flipping of plaquettes - namely the unflippable ones.

```python
import matplotlib.pyplot as plt
import sys
sys.path.append('../symmetry_stuff/')
from lattice_object import LatticeObject


lost_states = [15673952, 12368258, 8217412, 3076262, 13350953, 9972683, 5894413, 680687, 16096528, 10882802, 6804532, 3426262, 13700953, 8559803, 4408957, 1103263]

def draw_collection(states, filename=None, figsize=(24,10), L=[2,2,2]):
    n_cols = 4
    n_rows = len(states)//n_cols + (1 if len(states)%n_cols else 0)
    fig, axes = plt.subplots(n_rows, n_cols, subplot_kw={'projection': '3d'})
    fig.set_size_inches(*figsize)
    axes = axes.flatten()

    for i, c in enumerate(states):
        latt = LatticeObject(c, L)
        latt.draw(c, axis=axes[i], label='$|{:d}\\rangle$'.format(i+1))

    fig.tight_layout()
    if filename is not None:
        fig.savefig(filename)
        
draw_collection(sorted(lost_states), L=[2,2,2], figsize=(24,16))
```
