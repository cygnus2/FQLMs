# Comparison of diagonalization results.
These are plots to compare Debasish's and Lukas' results.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

db_datadir = '../c_code/ED_3d/'
lr_datadir = '../python_data/spectrum_compare/'
```

## 3D - 2x2x2

```python
db_bose = pd.read_csv(db_datadir+'boseLinks/evals.dat', names=['n', 'energy'], sep='\s+')
lr_bose = pd.read_csv(lr_datadir+'spectrum_bosons_2x2x2_lam0.00.dat', sep=',', names=['e_real', 'e_imag'])

db_fermi = pd.read_csv(db_datadir+'fermiLinks/evals.dat', names=['n', 'energy'], sep='\s+')
lr_fermi = pd.read_csv(lr_datadir+'spectrum_fermions_2x2x2_lam0.00.dat', sep=',', names=['e_real', 'e_imag'])

with plt.style.context('seaborn'):
    fig, ax = plt.subplots()
    
    cutoff = 50
    
    # Bosons.
    line = ax.plot(db_bose['energy'][:cutoff], label='Debasish (bosons, ED)')
    ax.plot(lr_bose['e_real'][:cutoff], label='Lukas (bosons, ARPACK)', marker='o', ls='', color=line[0].get_color())
    
    # Fermions       
    line = ax.plot(db_fermi['energy'][:cutoff], label='Debasish (fermions, ED)')
    ax.plot(lr_fermi['e_real'][:cutoff], label='Lukas (fermions, ARPACK)', marker='s', ls='', color=line[0].get_color())

    
    # ---
    
    ax.set_xlabel('n')
    ax.set_ylabel('e')
    ax.legend(loc='lower right')
    
#     fig.savefig('spectrum_2x2x2.png')
```

## 2D results

### 2x2

```python
db_bose = pd.read_csv('../c_code/CODES1F/lam0.0/Oflip_2x2.dat', names=['energy', '---'], sep='\s+')
lr_bose_full = pd.read_csv(lr_datadir+'FULL_spectrum_bosons_2x2_lam0.00.dat', sep=',', names=['e_real', 'e_imag'])
lr_fermi_full = pd.read_csv(lr_datadir+'FULL_spectrum_fermions_2x2_lam0.00.dat', sep=',', names=['e_real', 'e_imag'])


with plt.style.context('seaborn'):
    fig, ax = plt.subplots()
    
    cutoff = 6
    
    # Bosons.
    line = ax.plot(db_bose['energy'][:cutoff], label='Debasish (bosons, ED)')
    ax.plot(sorted(lr_bose_full['e_real'][:cutoff]), label='Lukas (bosons, ED)', marker='s', ls='', color=line[0].get_color())
    
    # Fermions.
    ax.plot(sorted(lr_fermi_full['e_real'][:cutoff]), label='Lukas (fermions, ED)', marker='o', ls='', color='red')
    
    
    ax.set_xlabel('n')
    ax.set_ylabel('e')
    ax.legend(loc='lower right')
```
