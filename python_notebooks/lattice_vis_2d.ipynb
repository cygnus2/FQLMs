```python
import matplotlib.pyplot as plt
import sys

sys.path.append('../symmetry_stuff/')
from lattice_object import LatticeObject

def draw_collection(states, filename=None, figsize=(24,10)):
    n_cols = 4
    n_rows = len(states)//n_cols + (1 if len(states)%n_cols else 0)
    fig, axes = plt.subplots(n_rows, n_cols, subplot_kw={'projection': '3d'})
    fig.set_size_inches(*figsize)
    axes = axes.flatten()

    for i, c in enumerate(states):
        latt = LatticeObject(c, [2,2,2])
        latt.draw(c, axis=axes[i], label='$|{:d}\\rangle$'.format(i+1))

    fig.tight_layout()
    if filename is not None:
        fig.savefig(filename)
        
draw_collection(contributions, filename='../../notes_overleaf/Notes/img/all_lattices.pdf')
```

### Draw some worldines.

```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)

        
def draw_worldline(latt, path, a=2, color='gray'):
    """ Draws a worldline for a specific particle.
    """ 
    pts = []
    for p in path:
        pt = np.array(p[:-1])*a
        d = p[-1]
        l = latt._get_drawing_links(pt, a)[d]
        
        shift = np.zeros(len(pt))
        shift[d] -= a/2
        pts.append(l+shift)
    
    for k in range(len(pts)-1):
        a = Arrow3D(
            [pts[k][0],pts[k+1][0]], 
            [pts[k][1],pts[k+1][1]], 
            [pts[k][2],pts[k+1][2]], 
            mutation_scale=20, lw=3, arrowstyle="-|>", color=color)
        latt.ax.add_artist(a)
```
