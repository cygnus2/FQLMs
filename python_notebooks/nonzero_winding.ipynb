# Maximally flippable states in non-zero winding sectors

```python
import sys

sys.path.append('../symmetry_stuff/')
from lattice_object import LatticeObject
```

We start by drawing a maximally flippables for 2x2x2.

```python
s0 = int('001110100011110001011100'[::-1],2)
latt = LatticeObject(s0, [2,2,2])
fig, ax = latt.draw(alpha=-0.1)


s1 = int('001100100001110001011100'[::-1],2)
latt = LatticeObject(s1, [2,2,2])
fig, ax = latt.draw(alpha=-0.1)
print("difference: {:024b}".format(s1^s0))
print("---")

s2 = int('001110100011100001001100'[::-1],2)
latt = LatticeObject(s2, [2,2,2])
fig, ax = latt.draw(alpha=-0.1)
print("difference: {:024b}".format(s2^s0))
```

Now, let's check how large the [1,0,0] winding sector actually is.

```python
import h5py as hdf 
import numpy as np

def read_states(state_file, ws='all-ws', merged=True):
    """ Read the GLS from file.
    """
    try:
        states = [] 
        if ws == 'all-ws':
            with hdf.File(state_file, 'r') as f:
                for ws in f:
                    if merged:
                        states += list(f[ws][...])
                    else:
                        states.append([ws, list(f[ws][...])])
        else:
            with hdf.File(state_file, 'r') as f:
                states = f[ws][...]

        return states

    except (OSError):
        return []
        
datadir = "../python_data/local_state_storage/"
all_states = np.sort(read_states(datadir+"winding_states_2x2x4.hdf5", ws='wx_6-wy_4-wz_2', merged=False))
print(f"Total # of states: {len(all_states)}")

# nflips = np.zeros(shape=(len(all_states)), dtype=np.int)
# for k, state in enumerate(all_states):
#     latt = LatticeObject(state, [2,2,4], quiet=True)
#     nflips[k] = latt.flippables()
    
# unflippables = all_states[nflips==0]
# print(f"# of unflippable states: {len(unflippables)}")
# print(f"# of other states: {len(all_states)-len(unflippables)}")
```
