# Comparison of diagonalization results.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 3D results.

```python
db_bose = pd.read_csv('../../ED_3d/boseLinks/evals.dat', names=['n', 'energy'], sep='\s+')
db_fermi = pd.read_csv('../../ED_3d/fermiLinks/evals.dat', names=['n', 'energy'], sep='\s+')
lr_full = pd.read_csv('../output/FULL_spectrum_winding_states_2x2x2.dat', sep=',', names=['e_real', 'e_imag'])
lr_sparse = pd.read_csv('../output/spectrum_winding_states_2x2x2.dat', sep=',', names=['e_real', 'e_imag'])

with plt.style.context('seaborn'):
    fig, ax = plt.subplots()
    
    cutoff = 20
    
    # Debasish ED.
    ax.plot(db_bose['energy'][:cutoff], label='Debasish (bosons, ED)')
    ax.plot(db_fermi['energy'][:cutoff], label='Debasish (fermions, ED)')
    
    # Lukas Lanczos.
    ax.plot(lr_full['e_real'][:cutoff], label='Lukas (bosons, ED)')
    ax.plot(lr_sparse['e_real'][:cutoff], label='Lukas (bosons, ARPACK)', marker='o', ls='')
    
    
    ax.set_xlabel('n')
    ax.set_ylabel('e')
    ax.legend(loc='lower right')
```
