# 3D spectra for the Gauss lattice

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Fermions

```python
db222_fermi = pd.read_csv('../../ED_3d/fermiLinks/evals.dat', names=['n', 'energy'], sep='\s+')
lr222_fermi = pd.read_csv('../output/SEQUENTIAL_spectrum_fermions_2x2x2_J-1.00.dat', sep=',', names=['e_real', 'e_imag'])
lr224_fermi = pd.read_csv('../output/SEQUENTIAL_spectrum_fermions_2x2x4_J-1.00.dat', sep=',', names=['e_real', 'e_imag'])

with plt.style.context('seaborn'):
    fig, ax = plt.subplots()
    
    cutoff = 100
    
    # Fermions       
    line = ax.plot(db222_fermi['energy'][:cutoff], label='Debasish (2x2x2, fermions, ED)')
    ax.plot(lr222_fermi['e_real'][:cutoff], label='Lukas (2x2x2, fermions, ARPACK)', marker='o', ls='', color=line[0].get_color())
    ax.plot(lr224_fermi['e_real'][:cutoff], label='Lukas (2x2x4, fermions, ARPACK)', marker='o', ls='')

    
    # ---
    
    ax.set_xlabel('n')
    ax.set_ylabel('e')
    ax.legend(loc='lower right')
    
    fig.savefig('spectrum_3D.png')
```

```python
db222_fermi = pd.read_csv('../../ED_3d/fermiLinks/evals.dat', names=['n', 'energy'], sep='\s+')
lr222_fermi = pd.read_csv('../output/SEQUENTIAL_spectrum_fermions_2x2x2_J-1.00.dat', sep=',', names=['e_real', 'e_imag'])
lr224_fermi = pd.read_csv('../output/SEQUENTIAL_spectrum_fermions_2x2x4_J-1.00.dat', sep=',', names=['e_real', 'e_imag'])

with plt.style.context('seaborn'):
    fig, ax = plt.subplots(1, 2, squeeze=False)
    fig.set_size_inches(18, 6)
    
    cutoff = 120
    
    # 2x2x2
    line = ax[0,0].plot(db222_fermi['energy'][:cutoff], label='Debasish (2x2x2, fermions, ED)')
    ax[0,0].plot(lr222_fermi['e_real'][:cutoff], label='Lukas (2x2x2, fermions, ARPACK)', marker='o', ls='', color=line[0].get_color())
    
    # 2x2x4
    ax[0,1].plot(lr224_fermi['e_real'][:cutoff], label='Lukas (2x2x4, fermions, ARPACK)', marker='o', ls='')

    
    # ---
    for a in ax.flatten():
        a.set_xlabel('n')
        a.set_ylabel('e')
        a.legend(loc='lower right')
    
#     fig.savefig('spectrum_3D.png')
```
